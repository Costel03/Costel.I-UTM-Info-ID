#### Unicode

Unicode este un standard internațional pentru reprezentarea caracterelor în format numeric, astfel încât orice simbol scris - litere, cifre, semne de punctuație, caractere speciale sau emoji are un code point - un numar intreg, de obicei exprimat in hexazecimal, care poate avea valori intre 0 pana la 1.114.111 (10FFFF).

Intervale

- U+0000 - U+007F → ASCII (basic Latin)
- U+0080 - U+FFFF → BMP (Basic Multilingual Plane, common scripts)
- U+10000 - U+10FFFF → supplementary (emoji, rare scripts)

Pentru a reprezenta toate code-pointurile din unicode sunt necesari maximum 21 de biti.

Reprezentarea in memorie a caracterelor unicode trebuie ales un pattern de biti, numit encoding. Din multe motive, exista mai multe moduri de reprezentare a code-pointurilor - utf8, utf16, utf32. Fiecare cu problemele sale.

- utf32 stocheaza fiecare caracter pe 32 de biti. Acest format permite toate code-pointurile dar este extrem de ineficient - majoritatea textului de pe un sistem informatic obisnuit este compus din caractere ASCII. Codul escape pentru acesta este \U
- utf8 stocheaza fiecare caracter intr-un numar variabil de biti. Caracterele ASCII de baza sunt stocate pe un singur octet (char), caracterele extinse (diacritice) sunt stocate de obicei in 2 octeti, emoji-urile si alte caractere in grafii mai neobisnuite in mai multi octeti (pana la 4). Nu exista secventa escape pentru utf8
- utf16 aparut initial ca encoding de lungime fixa, cand se credea ca 65536 code-points sunt suficiente pentru reprezentarea tuturor caracterelor existente. Odata cu introducerea caracterelor asiatice nu a mai fost suficient dar utf16 fusese deja adoptat in windows si java asa ca a fost si el convertit in encoding variabil. Secventa escape este \u dar nu poate prelucra decat spatiul de baza al utf16 (unul dintre cele 65536 caractere de baza). \u nu suporta caracterele unicode apartinand code-pointurilor superioare.

Exista online baza de date cu aceste caractere, care afiseaza toate modurile de
identificare a acestor caractere, denumire, variate tipuri de encoding, etc. https://unicode-table.com/en/24C3/

- |\N{Name} |	Numele caracterului Unicode. Poate reprezenta orice caracter unicode, atata timp cat are un nume.  |
- |\uxxxxxxxx |	Character unicode reprezentat pe 16 biti, poate reprezenta code-point-urile aflate in intervalul U+0000 to U+FFFF (Basic Multilingual plane) |
- |\Uxxxxxxxx	| Caracter unicode pe 32 de biti - oricare |
- |\000	| Character cu valoarea in octal - maximum 3 cifre, poate reprezenta doar caractere ASCII + ASCII extins |
- |\xhh	| Character cu valoarea hex - doua cifre |

Pentru reprezentarea oricarui caracter unicode pe baza de code-point, se foloseste functia chr()

Pentru listele si operatiile simple, comprehension este mai rapida, nu pentru ca foloseste alt algoritm ci pentru ca evita operatii de citire, evaluare, cautarea metodelor pe obiecte... etc.

```python
import timeit

def squares(size):
    result = []
    for number in range(size):
        result.append(number*number)
    return result

def squares_comprehension(size):
    return [number*number for number in range(size)]

print(timeit.timeit("squares(50)", "from __main__ import squares", number = 1000000))
print(timeit.timeit("squares_comprehension(50)", "from __main__ import squares_comprehension", number = 1000000))
```

### Copiere, atribuire

Atribuirea copiaza doar referinta. Se creaza o variabila de tip lista noua dar elementele din interior sunt aceleasi

```python
a = [1, 2, 3]
b = a
b[0] = 99
print(a)
```

Metoda copy() a unei liste poate fi folosita pentru copierea superficiala

```python
a = [1, 2, 3]
b = a.copy()
b[0] = 99
print(a)
```

Dar daca lista contine ca elemente alte liste, copierea superficiala va copia doar referintele acestora.

```python
a = [[1,2], [3,4]]
b = a.copy()
b[0][0] = 99
print(a)
```

Copierea produnda - deepcopy - face o copie completa

```python
import copy

a = [[1,2], [3,4]]
b = copy.deepcopy(a)
b[0][0] = 99
print(a)
```

## Tupluri

Un tuplu este o structură de date care reprezintă un grup fix de valori, ordonate și imutabile.

Inițial, tuplul nu a fost gândit ca o „colecție pentru utilizator”, ci ca o structură internă a interpreterului, folosită pentru a manipula eficient grupuri mici de valori: argumentele unei funcții, perechi (cheie, valoare), rezultate multiple etc. Pentru aceste situații, Python avea nevoie de un obiect simplu, de dimensiune fixă și fără modificări, iar tuple-ul îndeplinește exact acest rol.

În loc să ascundă această structură, Python a ales să o expună direct în limbaj. Astfel, ceea ce este folosit intern pentru organizarea datelor devine și un instrument disponibil programatorului.

Caracteristica esențială rămâne imutabilitatea: după creare, valorile nu se pot schimba. Din acest motiv, tuple-ul este stabil și potrivit pentru reprezentarea unor grupuri fixe de date care trebuie tratate ca o singură unitate.

## Seturi (multimi)

Un hash este o funcție care transformă un obiect într-un număr (de obicei un întreg), astfel încât acel obiect să poată fi localizat rapid într-o structură de date. Ideea este simplă: în loc să cauți un element parcurgând toate valorile (ca într-o listă), calculezi hash-ul lui și mergi direct în locul unde ar trebui să fie. Pentru ca acest mecanism să funcționeze corect, obiectele folosite trebuie să aibă un hash stabil, adică să nu se modifice în timp.

Pe baza acestui mecanism sunt construite set-urile din Python. Un set este o structură care stochează elemente folosind hash-ul lor, ceea ce permite verificări foarte rapide de tipul „există acest element?”. Din acest motiv, set-urile nu au ordine și nu permit duplicate: fiecare element este identificat prin hash și apare o singură dată. Astfel, set-ul nu este o listă mai limitată, ci o structură specializată pentru testarea rapidă a apartenenței și eliminarea automată a duplicatelor.

```python
mset = {"mere", "pere", "prune"}

print(mset)
print(type(mset))
print(dir(mset))
```

```python
mset = {"mere", "pere", "prune", "mere", "mere", "mere", "mere", "mere", "mere", "mere", "mere", "mere" }

print(mset)
```

```python
mlist = ["mere", "pere", "prune", "mere", "mere", "mere", "mere", "mere", "mere", "mere", "mere", "mere" ]

print(mlist)

unique = list(set(mlist))

print(unique)
```

setul nu poate contine obiecte mutabile pentru ca hash-ul acestora ar trebui fie sa se modifice odata cu modificarea lor. În principiu, ar fi posibil ca obiectele mutabile (cum sunt listele) să fie folosite pe post de chei și să li se refacă hash-ul atunci când se modifică. Însă asta ar necesita mecanisme suplimentare: obiectele ar trebui să notifice dicționarele că s-au schimbat, iar acestea ar trebui să-și reorganizeze intern toate pozițiile afectate. Rezultatul ar fi un sistem mai complex, mai lent și mai greu de înțeles. Din acest motiv, Python face o alegere de design clară: permite doar obiecte imutabile ca fiind hashabile, pentru a păstra performanța și comportamentul predictibil al structurilor bazate pe hashing.

### Operatii tipice cu seturi

#### Uniune

```python
set1 = {"fag", "frasin", "stejar", "alun"}
set2 = {"fag", "mesteacan", "balsa", "brad", "cais", "stejar"}

print(set1.union(set2))
```

#### Intersectie

```python
set1 = {"fag", "frasin", "stejar", "alun"}
set2 = {"fag", "mesteacan", "balsa", "brad", "cais", "stejar"}

print(set1.intersection(set2))
```

#### Diferenta

```python
set1 = {"fag", "frasin", "stejar", "alun"}
set2 = {"fag", "mesteacan", "balsa", "brad", "cais", "stejar"}

print(set1.difference(set2))
print(set2.difference(set1))
```

#### Diferenta simetrica

Diferenta simetrica dintre doua multimi retine toate elementele care nu fac parte din intersectia acestora

```python
set1 = {"fag", "frasin", "stejar", "alun"}
set2 = {"fag", "mesteacan", "balsa", "brad", "cais", "stejar"}

print(set1.symmetric_difference(set2))
print(set2.symmetric_difference(set1))
```

#### issuperset

Este setul curent supersetul unui alt set?

```python
set1 = {"fag", "frasin", "stejar", "alun", "brad", "cais", "stejar"}
set2 = {"brad", "cais", "stejar"}

print(set1.issuperset(set2))
print(set2.issuperset(set1))
```

#### difference_update, intersection_update si  symmetric_difference_update

difference_update:
 Inlatura dintr-un set elementele altui set

intersection_update:
  Adauga la set (daca nu exista deja) intersectia dintre setul curent si un alt set

symmetric_difference_update:
  Adauga la set (daca nu exista deja) diferenta simetrica dintre setul curent si un alt set

```python
set1 = {"fag", "frasin", "stejar", "alun", "brad", "cais", "stejar"}
set2 = {"brad", "cais", "stejar", "par"}

set1.difference_update(set2)
print(set1)
```

```python
set1 = {"fag", "frasin", "stejar", "alun", "brad", "cais", "stejar"}
set2 = {"brad", "cais", "stejar", "par"}

set1.intersection_update(set2)
print(set1)
```

```python
set1 = {"fag", "frasin", "stejar", "alun", "brad", "cais", "stejar"}
set2 = {"brad", "cais", "stejar", "par"}

set1.symmetric_difference_update(set2)
print(set1)
```

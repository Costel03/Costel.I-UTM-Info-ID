## Variabile si tipuri de date


Tipurile de date de baza sunt de doua feluri: singulare si colectii. Absolut toate variabilele în Python sunt obiecte definite în heap (malloc-ate) iar stack-ul conține doar referințe (pointeri) către acestea.

- Numele variabilelor scrise cu minuscule
- De evitat utilizarea cuvintelor cu semnificatie aparte (list, str)

În CPython, toate valorile de tip int sunt obiecte dinamice de tip PyLong. Nu există limită fixă de 32 sau 64 de biți; mărimea este limitată doar de memoria disponibilă.

Un int este stocat ca:

- un antet de obiect (informații despre tip și referințe),
- un câmp care indică semnul și numărul de segmente,
- un tablou de segmente numerice.

Fiecare segment conține 30 de biți utili (pe arhitecturi uzuale 64-bit). Valoarea totală a numărului este reconstruită ca sumă de segmente înmulțite cu puteri succesive ale lui 2^30.

Ordinea segmentelor este de la partea cea mai puțin semnificativă către cea mai semnificativă.

Numerele mici folosesc un singur segment. Numerele mai mari folosesc mai multe segmente. Nu există o limită fixă a valorii maxime sau minime; limita este memoria disponibilă.

Putem afla numarul de biti in care se stocheaza un segment se poate interoga folosind `sys.int_info.bits_per_digit`

```python
intvar = 659943549123443234

k = sys.int_info.bits_per_digit          # de obicei 30 pe 64-bit

base = 1 << k                            # 2^k
mask = base - 1                          # (2^k - 1)

segmente = []
x = intvar                               # copiem intvar in x
while x:
    segmente.append(x & mask)              # ia ultimii k biți si ii adauga la lista digits
    x >>= k                              # aruncă ultimii k biți

print("Biti pe digit: =", k)
print("Baza: =", bin(base))
print("nr segmente interne =", len(segmente))
print("segmente (cel mai putin semnificativ la inceput) =")
print(segmente)

# verificare: reconstruim numărul din segmente
reconstructed = 0
for i, d in enumerate(segmente):
    reconstructed += d * (base ** i)

print("reconstrucție OK?", reconstructed == intvar)
```

Exista situatii in care putem dori sa obtinem efectul rotunjirii in sus sau in jos, controlabil, la variabile care contin zecimale. Functia round() rootunjeste in sus sau in jos in functie de zecimale. Ceil si floor returneaza numere intregi. Putem scrie functii care "muta virgula" inainte si inapoi astfel incat sa putem utiliza functiile descrise mai sus (floor si ceil)

```python
def round_down(n, decimals=0):
    #calculeaza numarul cu care trebuie sa inmultim argumentul initial pentru a muta virgula la dreapta cu "decimals" zecimale
    multiplier = 10**decimals
    # aplica functia floor asupra numarului obtinut in urma mutarii virgulei, dupa care imparte la multiplicator pentru a muta virgula inapoi
    return math.floor(n * multiplier) / multiplier

def round_up(n, decimals=0):
    multiplier = 10**decimals
    return math.ceil(n * multiplier) / multiplier
```

```python
print('down: ',  round_down(1.2325, 2))
print('up: ',  round_up(1.2325, 2))


print('--------------------')
print(round(1.2325, 2))
```

Referintele catre numerele relativ mici sunt pastrate intr-un cache. La aplicarea unor operatii aritmetice, daca variabila revine la o valuare antrerioara, referinta va reveni la referinta anterioara. Aceasta se intampla doar la numerele mici iaar mecanismul se numeste "interning" (internalizare)

Pentru a observa functionarea referintelor si a mecanismului de "garbage collection", urmariti urmatorul bloc de cod, care genereaza cateva numere aleatoare intregi dincolo de cele din cache care sunt acum nemuritoare, genereaza (folosind o lista de liste) o serie variabila de referinte pentru fiecare dintre ele, afiseaza numarul de referinte, apoi goleste lista si afiseaza din nou numarul de referinte.

```python
import sys, random

# generam cateva numere intregi aleator
nums = [random.randint(1000, 1100) for _ in range(12)]

#initializam o lista care sa mentina un numar variabil de referinte
lista_generala = []

#in interiorul listei punem cate o sub-lista cu un numar variabil de referinte pentru fiecare
#dintre numerele generate
for x in nums:
    k = random.randint(1, 10)
    refs = [x] * k
    lista_generala.append(refs)
    print(x, "->", k, "refs:", refs)


print("\nNumarul de referinte:")
for x in nums:
    print("Numar:", x, "numarul de referinte:", sys.getrefcount(x))


# golim lista de toate elementele sale
lista_generala.clear()


print("\nNumarul de referinte dupa stergerea listei:")
for x in nums:
    print("Numar:", x, "numarul de referinte:", sys.getrefcount(x))
```

Devine clar astfel ca referintele nu sunt referinte catre variabile ci referinte catre valori. Acestea sunt in realitate obiectele stocate in memorie.

Putem folosi izoarea (slicing)

```python
string = 'abcdefghijklmnopqrstuvwxyz'


slice1 = string[10:]
slice2 = string[:10]     #de la caracterul 0 pana la caracterul 10 (exclusiv)
slice3 = string[4:8]     #de la caracterul 5 pana la caracterul 9 (exclusiv)
slice4 = string[4:18:3]  #de la caracterul 5 pana la caracterul al 18-lea cu pasul 3
slice5 = string[::3]     #de la început pana la final cu pasul 3
slice6 = string[::-1]    #pasul negativ baleiaza de la dreapta la stanga

print('slice1 ', slice1)
print('slice2 ', slice2)
print('slice3 ', slice3)
print('slice4 ', slice4)
print('slice5 ', slice5)
print('slice6 ', slice6)
```

Unicode

Unicode este un standard internațional pentru reprezentarea caracterelor în format numeric, astfel încât orice simbol scris - litere, cifre, semne de punctuație, caractere speciale sau emoji are un **code point** - un numar intreg, de obicei exprimat in hexazecimal, care poate avea valori intre 0 pana la 1.114.111 (10FFFF). Pentru a reprezenta toate code-pointurile din unicode sunt necesari maximum 21 de biti.

Reprezentarea in memorie a caracterelor unicode trebuie ales un pattern de biti, numit encoding. Din multe motive, exista mai multe moduri de reprezentare a code-pointurilor - utf8, utf16, utf32. Fiecare cu problemele sale.

- **utf32** stocheaza fiecare caracter pe 32 de biti. Acest format permite toate code-pointurile dar este extrem de ineficient - majoritatea textului de pe un sistem informatic obisnuit este compus din caractere ASCII. Codul escape pentru acesta este \U
- **utf8** stocheaza fiecare caracter intr-un numar variabil de biti. Caracterele ASCII de baza sunt stocate pe un singur octet (char), caracterele extinse (diacritice) sunt stocate de obicei in 2 octeti, emoji-urile si alte caractere in grafii mai neobisnuite in mai multi octeti (pana la 4). Nu exista secventa escape pentru utf8
- **utf16** aparut initial ca encoding de lungime fixa, cand se credea ca 65536 code-points sunt suficiente pentru reprezentarea tuturor caracterelor existente. Odata cu introducerea caracterelor asiatice nu a mai fost suficient dar utf16 fusese deja adoptat in windows si java asa ca a fost si el convertit in encoding variabil. Secventa escape este \u dar nu poate prelucra decat spatiul de baza al utf16 (unul dintre cele 65536 caractere de baza). \u nu suporta caracterele unicode apartinand code-pointurilor superioare.   

Exista online baza de date cu aceste caractere, care afiseaza toate modurile de identificare a acestor caractere, denumire, variate tipuri de encoding, etc.

### Dicționare (dict)

Dictionarele sunt un tip de date care poate stoca informatii conectate. Structura acestora este de perechi cheie - valoare.

Membrii dictionarului nu pastreaza nici o ordine - extragerea informatiei poate conduce la alta ordine decat adaugarea ei (pana la Python 3.7)

````python
md = {'one' : 12, 'two': 32, 'three': 87}
print(md)
print(type(md))

print(dir(md))
````

Sortarea dupa valoare este una dintre cele mai importante operatii care pot fi necesare in lucrul cu dictionare.

#### Dictionare: comprehension

Ca si in cazul listelor, dictionary comprehension reprezinta o serie de tehnici care tin de sintaxa limbajului de a obtine in foarte putine linii de cod ceea ce ar putea fi obtinut in blocuri semnificativ mai mari

Dictionary comprehension implica initializarea dictionarelor folosind blocuri de cod in loc de perechi cheie - valoare


structura generala este astfel:

in loc sa scriem dictionar_nou = {'cheie': 'valoare'} inlocuim una dintre cele doua (sau pe ambele) cu un bloc de cod.

dictionar_nou = {cheie: for ... if ...}

De exemplu, daca dorim sa inversam cheile cu valorile:


````python
swapped_dict = {value: key for key, value in masini.items()}

print(swapped_dict)
````

Adaugarea unei unitati la toate valorile unui dictionar:

````python
mai_multe_masini = {masina: numar + 1 for masina, numar in masini.items()}
print(mai_multe_masini)
````

evident, oricand se poate recurge la metoda clasica:

````python
mai_multe_masini = {}

for masina, numar in masini.items():
  mai_multe_masini[masina] = numar + 1

print(mai_multe_masini)
````

expresiile pot fi mai complicate - ele pot contine conditionale
De exemplu, sa presupunem ca vrem sa mai adaugam 1 la valorile din mai_multe_masini dar doar pentru cele care au valoarea mai mica de 10.

Varianta clasica arata astfel:

````python
mm_masini = {}

for masina, numar in mai_multe_masini.items():
  if numar < 10:
    mm_masini[masina] = numar + 1
  else:
    mm_masini[masina] = numar

print(mm_masini)
````

varianta in care folosim dictionary comprehension:

````python
mm_masini_dc = {masina: numar + 1 if numar < 10 else numar for masina, numar in mai_multe_masini.items()}
print (mm_masini_dc)
````

un alt exemplu: conversia de la farenheit la celsius:

vom itera peste tot dictionarul farenheit si vom folosi la fiecare item formula de conversie:

(5/9) * (F-32)

cheile raman nemodificate, valorile se modifica conform formulei

````python
fahrenheit = {'t1': -30,'t2': -20,'t3': -10,'t4': 0}

celsius = {k:(float(5)/9)*(v-32) for (k,v) in fahrenheit.items()}

print(celsius)
````

Putem adauga oricate conditii la dictionary comprehension

````python
print(mai_multe_masini)

mm_masini_dc_2 = {masina: numar for masina, numar in mai_multe_masini.items() if numar >4  if masina.startswith('l') }
print(mm_masini_dc_2)
````

observam ca aparitia lui else modifica sintaxa impingand for-ul pe ultimul loc. Daca dorim sa folosim mai multe if-uri for-ul trebuie sa vina inaintea acestora si nu mai putem folosi else

Cache al rezultatelor unei functii

````python
import time

cache = {}

def compute_cost(page, category, price_min, price_max):
    time.sleep(2)
    return page * 10 + (price_max - price_min) #expresie oarecare


calls = [
    {"page": 1, "category": "books",       "price_min": 10, "price_max": 100},
    {"category": "electronics", "price_max": 300, "price_min": 50, "page": 2},
    {"price_max": 100, "page": 1, "category": "books", "price_min": 10},   # repeat (different order)
    {"price_min": 5, "category": "toys", "page": 3, "price_max": 80},
    {"category": "clothes", "page": 1, "price_max": 120, "price_min": 20},
    {"price_min": 50, "page": 2, "price_max": 300, "category": "electronics"},   # repeat (different order)
    {"page": 4, "price_max": 90, "category": "books", "price_min": 15},
    {"category": "garden", "price_min": 30, "price_max": 200, "page": 5},
    {"category": "toys", "price_max": 80, "price_min": 5, "page": 3},    # repeat (different order)
    {"price_min": 25, "category": "sports", "page": 1, "price_max": 150},
]

for i, params in enumerate(calls, start=1):
    key = frozenset(params.items())

    start = time.time()

    if key in cache:
        result = cache[key]
        source = "cache"
    else:
        result = compute_cost(**params)
        cache[key] = result
        source = "computed"

    elapsed = time.time() - start

    print(f"call {i}: {source:8} -> result={result}, time={elapsed:.3f}s")
````

In python 2, range genera toate elementele la rulare. Acest comportament determina utilizari semnificative ale memoriei iar daca range era utilizat doar pentru a itera printr-un set de date existent, acea memorie era consumata degeaba. In Python2 exista functia xrange care genera elementele unul cate unul. In python 3, range se comporta ca xrange din python 2 asa ca xrange nu mai este necesar.

Prin comparare cu generarea unei liste complete de numere intregi, range este mult mai rapid

````python
import time
import tracemalloc

N = 10_000_000
INDEX = 9_000_000

# -------- LIST --------
tracemalloc.start()
t0 = time.time()

lst = list(range(N))              # construieste intreaga lista
value_list = lst[INDEX]           # acceseaza un index mare

t1 = time.time()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print("LIST")
print("value:", value_list)
print("time:", round(t1 - t0, 3), "s")
print("memory (MB):", round(peak / 1024 / 1024, 2))


# -------- RANGE --------
tracemalloc.start()
t0 = time.time()

r = range(N)                     # nu se creaza lista
value_range = r[INDEX]           # accesarea unui index mare din range

t1 = time.time()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print("\nRANGE")
print("value:", value_range)
print("time:", round(t1 - t0, 3), "s")
print("memory (MB):", round(peak / 1024 / 1024, 2))
````

#### Continue

Continue "sare" peste un ciclu al iteratiei la urmatorul. In urmatorul exemplu, cand i devine eval cu 3, instructiunea "continue" sare peste print(i) si determina solicitarea urmatoarei valori a lui i (4) de la iterator.

#### for ... else??

Python permite adaugarea unui else la o bucla for. Aceasta se va executa la terminarea iteratiei, cu exceptia cazului in care s-a folosit conditia break.

````python
for i in range(4):
  print(i)
else:
  print('stop')

print('post loop')
````

O utilizare frecventa a acestui else este in cazurile in care avem o serie de numere pe care dorim sa le testam iar testul implica o bucla for interna. Un exemplu frecvent este determinarea unui set de numere prime, intr-o lista de numere date sau un anumit range.

Bucla for exterioara (n) trece prin toate numerele care vor fi testate

Bucla interioara (care face parte din test) trebuie sa verifice daca numarul curent se divide exact cu oricare dintre numerele mai mici decat el. Daca aceasta bucla interna se termina fara break, putem folosi un else pentru a anunta numarul prim.

````python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'equals', x, '*', n//x)
            break
    else:
        # loop fell through without finding a factor
        print(n, 'is a prime number')
````

O alta utilizare este in cautari a unor elemente in liste sau altfel de colectii. In codul urmator else se va executa doar daca numarul cautat nu a fost gasit

````python
numere = [1, 3, 5, 7]
search_for = 4

for num in numere:
    if num == search_for:
        print(f"Found {search_for}!")
        break
else:
    print(f"{search_for} not found in the list.")
````

### zip

functie care preia mai multe iterabile si construieste tupluri din ele. Denumirea sugereaza un fermoar.

````python
a = [1, 2, 3]
b = ["a", "b", "c"]

for x, y in zip(a, b):
    print(x, y)
````

````python
names = ["Ana", "Dan", "Mihai"]
scores = [10, 8, 9]
cities = ["București", "Cluj", "Iași"]

for name, score, city in zip(names, scores, cities):
    print(name, score, city)
````

### match

Match este echivalentul Python al declaratiei "switch" din alte limbaje de programare

````python
status = 400

match status:
  case 200:
    print("all ok")
  case 400:
    print ("Bad request")
  case 404:
    print("Not found")
  case 418:
    print("I'm a teapot")
  case _:
    print("Something's wrong with the internet")
````

Underscore (_) este folosit pentru a capta situatiile in care nu exista un match anterior. Acesta este numit "irrefutable pattern" si trebuie pozitionat pe ultimul loc, altfel capatam eroarea:

SyntaxError: wildcard makes remaining patterns unreachable

````python
status = 400

match status:
  case _:
    print("Something's wrong with the internet")
  case 200:
    print("all ok")
  case 400:
    print ("Bad request")
  case 404:
    print("Not found")
  case 418:
    print("I'm a teapot")
````

match poate fi folosit si pentru compararea structurii

````python
data = (10, 20)

match data:
    case (x, y):
        print(x, y)
````

````python
point = (0, 5)

match point:
    case (0, y):
        print("on Y axis:", y)
````

````python
data = [1, 2, 3, 4]

match data:
    case [first, *rest]:
        print(first, rest)
````

````python
user = {"name": "Ana", "age": 25}

match user:
    case {"name": name}:
        print(name)
````

conditiile pot include si comparatii

````python
items = [
    (10, 20),
    (7, 3),
    (5, 5),
    (1, 9),
    (100,),
    "abc",
    (8, 2),
]

for data in items:
    match data:
        case (x, y) if x < y:
            print(data, "-> increasing")
        case (x, y) if x > y:
            print(data, "-> decreasing")
        case (x, y):
            print(data, "-> equal")
        case _:
            print(data, "-> not a pair")
````

### Modul de transmitere al argumentelor

Noțiunile de argument și parametru sunt adesea folosite interschimbabil, dar există o diferență:

 - parametrii apar în definiția funcției
 - argumentele sunt valorile transmise la apelul funcției

În multe limbaje de programare se discută despre două moduri de transmitere a argumentelor:

 - by value (prin valoare)
 - by reference (prin referință)

În Python, această distincție nu se aplică în mod direct.

În Python, toate variabilele sunt nume care fac referire la obiecte din memorie

La apelul unei funcții:
 - parametrii devin nume locale care se leagă la aceleași obiecte ca argumentele
 - Nu se creează automat copii ale obiectelor.

Anumite tipuri de date sunt imutabile:

 - numere (int, float, etc.)
 - șiruri de caractere
 - tupluri
 - frozenset

Altele sunt mutabile:

 - liste
 - dicționare
 - seturi

Diferența de comportament apare din mutabilitatea obiectelor, nu din modul de transmitere:

 - obiectele imutabile nu pot fi modificate → orice “modificare” creează un obiect nou
 - obiectele mutabile pot fi modificate → modificările sunt vizibile și în exterior

De aceea:

 - pentru tipurile imutabile, comportamentul pare similar cu by value
 - pentru tipurile mutabile, comportamentul pare similar cu by reference


Daca trimitem un numar intreg catre o functie dar nu il modificam in interiorul acesteia, referintele raman aceleasi.

````python
def changeme1(vr):
  print(vr, id(vr))
  vr = 15
  print(vr, id(vr))

a = 30
print(a, id(a))
changeme1(a)
print(a, id(a))
````

daca trimitem catre functie o lista care contine un singur numar intreg si il modificam in lista, modificarile se vor mentine dupa iesirea din functie, pentru ca lista este un tip de date mutabil.

````python
def changeme2(vrl):
  print(vrl, id(vrl))
  vrl[0] = 15
  print(vrl, id(vrl))

a = [30]
print(a, id(a))
changeme2(a)
print(a, id(a))
````

````python
def changeme3(vrl2):
  print(vrl2, id(vrl2))
  vrl2.append(15)
  print(vrl2, id(vrl2))

a = [30]
print(a, id(a))
changeme3(a)
print(a, id(a))
````

Exista vreo posibilitate de a trimite o lista astfel incat sa nu fie afectata de modificarile in interiorul functiei, emuland tipul de trimitere "by value"? Pentru a obtine acest efect trebuie sa facem undeva o copie a listei, aceasta trebuie sa fie completa (deep copy) daca e nevoie iar copia trebuie facuta explicit din cod.

Toate operatiile native din Python de copiere a listelor sunt shalow copy. Acestea se pot folosi pentru trimiterea unei liste by_value dar trebuie sa nu uitam ca acea copiere este shallow_copy

- slicing: changeme3(a[:])
- list(): changeme3(list(a))
- list.copy(): changeme3(a.copy())

toate cele trei variante au acelasi efect:

````python
a = [30]
print(a, id(a))
changeme2(a.copy()) # nu modificam lista primita ca argument ci o copie
print(a, id(a))
````

dar daca lista contine o referinta (de exemplu catre o alta lista) si modificarea afecteaza tocmai acea referinta, modificarea va ramane dupa terminarea executiei functiei.

````python
def changeme4(vr):
  print(vr, id(vr))
  vr[1][0] = 4
  print(vr, id(vr))


a = [30, [15]]
print(a, id(a))
changeme4(a.copy())
print(a, id(a))
````

Exista posibilitatea de a incurca borcanele semnificativ de rau, trimitand o lista cu referinte sub forma de shallow copy si facand modificari atat in zona copiata cat si in zona de referinte. O parte din modificari vor ramane dupa incheierea executiei functiei, o parte nu.

````python
def changeme4(vr):
  print(vr, id(vr))
  vr[1][0] = 4
  vr.append(8)
  print(vr, id(vr))


a = [30, [15]]
print(a, id(a))
changeme4(a.copy())
print(a, id(a))
````

In exemplul urmator se pierde complet referinta catre sublista [15,16,17] chiar daca atribuirea se face in interiorul functiei, chiar daca se foloseste copy pentru ca referinta interioara ramane aceeasi

````python
def changeme5(vr):
  print(vr, id(vr))
  vr[2] = 4
  print(vr, id(vr))


a = [30, 32,[15,16,17],32]
print(a, id(a))
changeme5(a.copy())
print(a, id(a))
````

Pentru evitarea acestor probleme se poate folosi modulul deepcopy - acesta va duplica si referintele din lista astfel incat inlocuirea lor in interiorul functiei nu se va mentine la iesirea

````python
import copy

def changeme6(vr):
  print(vr, id(vr))
  vr[1][0] = 4
  vr.append(8)
  print(vr, id(vr))


a = [30, [15]]
print(a, id(a))
changeme6(copy.deepcopy(a))
print(a, id(a))
````

#### Functii cu numar variabil de parametrii de tip keyword

O functie poate astepta un numar variabil de parametrii de tip keyword.

````python
def calculate_total_cost(price, **kwargs):
    tax_rate = kwargs.get('tax_rate', 0.05)
    discount = kwargs.get('discount', 0)
    shipping_cost = kwargs.get('shipping_cost', 5.0)

    total_cost = price + (price * tax_rate) - discount + shipping_cost
    return total_cost

# Example usage
price = 100.0
total_cost = calculate_total_cost(price, tax_rate=0.08, discount=10.0, shipping_cost=8.0)
print(f"Total cost: {total_cost}")
````
